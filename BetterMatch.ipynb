```python
#creating a better match.csv
import pandas as pd
```

```python
matchDf = pd.read_pickle('match_CLEANED.pkl')
matchDf['year'] = pd.DatetimeIndex(matchDf['date']).year
matchDf.head()
```

```python
team_attributesDf = pd.read_pickle('team_attributes_CLEANED.pkl')
team_attributesDf['year'] = pd.DatetimeIndex(team_attributesDf['date']).year
team_attributesDf.head()
```

```python
#for using sql join to make things easier
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
```

```python
matchDf.to_sql('match', con=engine)
team_attributesDf.to_sql('team_attributes', con=engine)
```

```python
COMMAND = "SELECT match.match_id, match.home_team_id, match.away_team_id, match.date, match.home_team_goal, match.away_team_goal, home_team.buildUpPlaySpeed, home_team.buildUpPlayDribbling, home_team.buildUpPlayPassing, home_team.chanceCreationPassing, home_team.chanceCreationCrossing, home_team.chanceCreationShooting, home_team.defencePressure, home_team.defenceAggression, home_team.defenceTeamWidth, away_team.buildUpPlaySpeed, away_team.buildUpPlayDribbling, away_team.buildUpPlayPassing, away_team.chanceCreationPassing, away_team.chanceCreationCrossing, away_team.chanceCreationShooting, away_team.defencePressure, away_team.defenceAggression, away_team.defenceTeamWidth FROM match INNER JOIN team_attributes AS home_team ON home_team.team_id = match.home_team_id AND home_team.year = match.year INNER JOIN team_attributes AS away_team ON away_team.team_id = match.away_team_id AND away_team.year = match.year"
complete_matchDf = pd.DataFrame(engine.execute(COMMAND))
```

```python
proper_column_name = [
    'match_id',
    'home_team_goal',
    'away_team_goal',
    'date',
    'home_team_goal',
    'away_team_goal',
    'home_team_build_up_play_speed',
    'home_team_build_up_play_dribbling',
    'home_team_build_up_play_passing',
    'home_team_chance_creationg_passing',
    'home_team_chance_creation_crossing',
    'home_team_chance_creation_shooting',
    'home_team_defence_pressure',
    'home_team_defence_aggression',
    'home_team_defence_team_width',
    'away_team_build_up_play_speed',
    'away_team_build_up_play_dribbling',
    'away_team_build_up_play_passing',
    'away_team_chance_creationg_passing',
    'away_team_chance_creation_crossing',
    'away_team_chance_creation_shooting',
    'away_team_defence_pressure',
    'away_team_defence_aggression',
    'away_team_defence_team_width',
]
```

```python
complete_matchDf.columns = proper_column_name
```
